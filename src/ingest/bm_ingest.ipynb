```python

# Simplified Regex for N-Quads tokens
# 1: IRIs, 2: BNodes, 3: Literals
TOKEN_PATTERN = r'(<[^>]+>|_:[\w\d]+|"(?:\\.|[^"\\])*"(?:@[a-zA-Z-]+|\^\^<[^>]+>)?)'


def parse_nq_line(line):
    
    line = line.strip()
    
    if not line or line.startswith("#"):
        return None

    # Extract tokens and remove the trailing dot
    tokens = re.findall(TOKEN_PATTERN, line)
    
    if len(tokens) < 3:
        return None # Invalid line
    
    subject = tokens[0]
    predicate = tokens[1]
    obj = tokens[2]
    graph = tokens[3] if len(tokens) > 3 else "default"
    quad = {
        "subject" : subject,
        "predicate":predicate,
        "object":obj,
        "graph":graph
    }
    
    return quad
        
        
print("ngparser")
```

```python

subjects_dict = {}

def explore_data(nq_file):
    
    print(f"Starting exploration of: {nq_file}")
    
    file_size = os.path.getsize(nq_file)
    
    with open(nq_file, 'r', encoding='utf-8') as f:
        with tqdm(total=file_size, unit='B', unit_scale=True, desc="Data Progress") as pbar:
            for line_no, line in enumerate(f, 1):

                
                try:
                    quad = parse_nq_line(line)
                    pbar.update(len(line.encode('utf-8')))
                    
    #                 if quad:
    #                 subjects_dict[quad["subject"]] = quad
        
        
                except Exception as e: 
                    return f"Error parsing line {line_no}: {e}"

        
    return "\n\n\n done \n\n\n"
    
    
res = explore_data(FULL_NQ_FILE)
print(res)
```
